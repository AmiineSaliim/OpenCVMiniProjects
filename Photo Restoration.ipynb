```python
import cv2 

import matplotlib.pyplot as plt

import numpy as np
```

```python
image = cv2.imread('./Images/Abraham.jpg')

image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)

plt.imshow(image)

plt.show()
```

```python
marked_damages = cv2.imread('./Images/Mask.jpg')

marked_damages = cv2.cvtColor(marked_damages , cv2.COLOR_BGR2RGB)

plt.imshow(marked_damages)

plt.show()
```

```python
ret, thresh1 = cv2.threshold(marked_damages, 254, 255, cv2.THRESH_BINARY)

plt.imshow(thresh1)

plt.show()
```

```python
kernel = np.ones((7,7), np.uint8)

mask = cv2.dilate(thresh1, kernel, iterations = 1)

plt.imshow(mask)

plt.show()
```

```python
cv2.imwrite('./Images/Abraham_Mask.png', mask)

mask = cv2.cvtColor(mask , cv2.COLOR_BGR2GRAY)

restored = cv2.inpaint(image , mask , 3 , cv2.INPAINT_NS)

plt.imshow(restored)

plt.show()

cv2.imwrite('./Images/Abraham_Restored.png' , restored)
```
